```python
# Importing libraries
import pandas as pd
from rdkit import Chem, DataStructs
from rdkit.Chem import AllChem
import numpy as np
import matplotlib.pyplot as plt

# Loading the dataset as pandas dataframe
df = pd.read_csv('compounds.csv', delimiter=';')
df.set_index("name", inplace=True)
display(df)
```

```python
# creating column with computed fingerprints
fpgen = AllChem.GetRDKitFPGenerator()

df["fingerprint"] = [fpgen.GetFingerprint(Chem.MolFromSmiles(x)) for x in df["smiles"]]
display(df)
```

```python
# reference compound
ref_name = "Levopoda"
ref_smiles = "C1=CC(=C(C=C1CC(C(=O)O)N)O)O"
ref_fingerprint = fpgen.GetFingerprint(Chem.MolFromSmiles(ref_smiles))

# creating column with Tanimoto similarity measure to reference compound
df["tanimoto"] = [DataStructs.TanimotoSimilarity(x,ref_fingerprint) for x in df["fingerprint"]]
display(df)
```

```python
# sorting and displaying top 3 hits
df.sort_values(by="tanimoto", ascending=False).head(3)
```

```python
# adding columns for Dice, Cosine, Sokal, Euclidian distance
df["dice"] = [DataStructs.DiceSimilarity(x,ref_fingerprint) for x in df["fingerprint"]]
df["cosine"] = [DataStructs.CosineSimilarity(x,ref_fingerprint) for x in df["fingerprint"]]
df["sokal"] = [DataStructs.SokalSimilarity(x,ref_fingerprint) for x in df["fingerprint"]]
df["euclidian"] = [np.linalg.norm(np.array(x, dtype=int) - np.array(ref_fingerprint, dtype=int)) for x in df["fingerprint"]]
display(df.sort_values(by="tanimoto", ascending=False))
```

```python
# Graphical representation

# normalizing and inverting euclidian distance
df["euclidian_normalized_inverted"] = (1 - ((df["euclidian"]-df["euclidian"].min())/(df["euclidian"].max()-df["euclidian"].min())))

# selecting only columns with similarities
colums_measures = ["tanimoto", "dice", "cosine", "sokal", "euclidian_normalized_inverted"]
df_measures = df[colums_measures].sort_values(by="tanimoto", ascending=False)

# plotting the results
plt.figure(figsize=(16, 5))
for column in df_measures.columns:
    plt.plot(df_measures.index, df_measures[column], marker='o', label=column)

plt.title("Similarity Measures")
plt.xlabel("Molecule Name") 
plt.ylabel("Similarity Score") 
plt.xticks(ticks=range(len(df_measures.index)), labels=df_measures.index, rotation=45, ha='right')
plt.legend()
plt.grid(True)

plt.show()
```
